```python
import pandas as pd
import numpy as np


#declaro la ruta
ruta = r"C:\Users\rikar\OneDrive\Documents\Gitlab\TranquiTest\csv"
outpout =  r"C:\Users\rikar\OneDrive\Documents\Gitlab\TranquiTest\csv clean files"

class data_wrangling():
    
   
    #inicializo variables
    def __init__(self):
        self.__deliveries = pd.read_csv(f"{ruta}\\deliveries.csv")
        self.__users = pd.read_csv(f"{ruta}\\users.csv")
        self.__segments = pd.read_csv(f"{ruta}\\segments.csv")
      
        
        
    #exploro datos para verificar si hay datos nulos
    def exploring_data(self):
        
        
        nullable_val_deliveries = self.__deliveries.isnull().sum()
        nullable_val_users = self.__users.isnull().sum()
        nullable_val_segments = self.__segments.isnull().sum()
        
        return nullable_val_deliveries, nullable_val_users,nullable_val_segments
    
    #exploro los tipos de datos
    def exploring_types(self):
        
        deliveries_types= self.__segments.dtypes
        users_types= self.__users.dtypes
        segments_types= self.__segments.dtypes
        return deliveries_types,users_types , segments_types
        
    #formateo las columnas necesarias de los dataframes
    def formating_columns(self):
        
        self.__segments.loc[:,'fecha_evolucion'] = pd.to_datetime(self.__segments.fecha_evolucion)
        return self.__segments
    
    #primera exploracion para verificar la calidad de los datos
    def finding_duplicates(self):
        
        segments = self.__segments.describe(include='all',datetime_is_numeric=True )
        users = self.__users.describe(include='all' )
        deliveries  = self.__deliveries.describe(include='all' )
        return segments,users,deliveries
            
    #verificando ,eliminando duplicados y nulos y exportando la data- - segments
    def exploring_duplicates_segments(self):
        
        clean_segments = pd.DataFrame()
        
        if self.__segments.isnull().sum().sum()==0:
            self.__segments= self.__segments[~self.__segments.duplicated(subset=['fecha_evolucion','document'],keep=False)].drop_duplicates()
            self.__segments.to_csv(f"{outpout}\\segments.csv",index=False)
            return self.__segments

        if self.__segments.isnull().sum().sum() >0:
            self.__segments= self.__segments[~self.__segments.duplicated(subset=['fecha_evolucion','document'],keep=False)].drop_duplicates().dropna()
            self.__segments.to_csv(f"{outpout}\\segments.csv",index=False)
            return self.__segments

        
        #verificando ,eliminando duplicados y nulos y exportando la data- - users
    def exploring_duplicates_users(self):
            
        if  self.__users.isnull().sum().sum()==0:
            self.__users = self.__users[~self.__users.duplicated(subset=['user_id','document'],keep=False)].drop_duplicates()
            self.__users.to_csv(f"{outpout}\\users.csv",index=False)
            return self.__users 

        if self.__users.isnull().sum().sum()>0:
            self.__users  = self.__users [~self.__users .duplicated(subset=['user_id','document'],keep=False)].drop_duplicates().dropna()
            self.__users.to_csv(f"{outpout}\\users.csv",index=False)
            return self.__users 
    
       #verificando ,eliminando duplicados y nulos y exportando la data- deliveries
    def exploring_duplicates_deliveries(self):
    
        
        if  self.__deliveries.isnull().sum().sum()==0 :
            self.__deliveries = self.__deliveries[~self.__deliveries.duplicated(keep=False)].drop_duplicates()
            self.__deliveries.to_csv(f"{outpout}\\deliveries.csv",index=False)
            return self.__deliveries
        
        
        if self.__deliveries.isnull().sum().sum()>0:
            self.__deliveries = self.__deliveries[~self.__deliveries.duplicated(keep=False)].drop_duplicates().dropna(subset=['user_id'])
            self.__deliveries.to_csv(f"{outpout}\\deliveries.csv",index=False)
            return self.__deliveries
        
    
    
        
```
